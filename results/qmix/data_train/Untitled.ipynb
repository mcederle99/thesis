```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time_steps = np.load('qmix2_time_steps.npy')
eval_rewards = np.load('qmix2_eval_rewards.npy')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, label="1")

N = len(eval_rewards)
print(N)
running_avg = np.empty(N)
for t in range(N):
    running_avg[t] = np.mean(
            eval_rewards[max(0, t-9):(t+1)])

ax.plot(time_steps, eval_rewards, color="C0")
ax.set_xlabel("Training Steps", color="C0")
ax.set_ylabel("QMIX Score", color="C0")
ax.tick_params(axis='x', colors="C0")
ax.tick_params(axis='y', colors="C0")
plt.grid(True, axis='y', linestyle=':')
plt.show()
```
