```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.load('maddpg_2_steps.npy')
maddpg_scores = np.load('maddpg_2_scores.npy')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, label="1")

N = len(maddpg_scores)
print(N)
running_avg = np.empty(N)
for t in range(N):
    running_avg[t] = np.mean(
            maddpg_scores[max(0, t-20):(t+1)])

ax.plot(x, running_avg, color="C0")
ax.set_xlabel("Training Steps", color="C0")
ax.set_ylabel("Avg. Return MADDPG Convolutional", color="C0")
ax.tick_params(axis='x', colors="C0")
ax.tick_params(axis='y', colors="C0")
plt.grid(True, axis='y', linestyle=':')
plt.show()
```
